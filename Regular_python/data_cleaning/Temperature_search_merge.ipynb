```python
import requests
import numpy as np
import pandas as pd
from pprint import pprint
import json
from urllib.request import urlopen, quote
```

```python
# Import data to achieve location value
Loc = pd.read_csv("Data_Merge.csv")
```

```python
Latitude = Loc.Latitude
Longitude = Loc.Longitude
LOC = pd.DataFrame()
LOC['Latitude'] = Latitude
la = LOC['Latitude']
LOC['Longitude'] = Longitude
lo = LOC['Longitude']
```

```python
LOC.head(2)
```

```python
def temperature(data):
    api_address = 'http://api.openweathermap.org/data/2.5/weather?&appid=37eb773813f7219dab7d60cec5dfa6d6'
    #Latitude_l = input("Lat :")
    #Longitude_l = input("Lon :")
    latitude_l = data[0]
    longitude_l = data[1]

    url = api_address + "&lat=" + str(latitude_l) + "&lon=" + str(longitude_l)
    #data = requests.get(url)
    #res = data.json()
    req = urlopen(url)
    res = req.read().decode()
    temp = json.loads(res)
    temper = temp['main']['feels_like']
    return temper
```

```python
data_list = []
for i in range(len(la)):
    row_list = []
    row_list.append(la[i])
    row_list.append(lo[i])
    data_list.append(row_list)
```

```python
temp_data = []
for i in data_list:
    temp_data.append(temperature(i))
```

```python
Loc['Temperature_03_12'] = temp_data
```
